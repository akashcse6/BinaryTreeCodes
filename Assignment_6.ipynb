# Task 1

```python
class Student:
    id = 0

    def __init__(self, name, department, age, cgpa):
        self.name = name
        self.department = department
        self.age = age
        self.cgpa = cgpa
        Student.id += 1

    def get_details(self):
        print("ID:",Student.id)
        print("Name:",self.name)
        print("Department",self.department)
        print("Age:",self.age)
        print("CGPA:",self.cgpa)

    @classmethod
    def from_String(cls, string):
        name, department, age, cgpa = string.split("-")
        return cls(name, department, age, cgpa)

s1 = Student("Samin", "CSE", 21, 3.91)

s1.get_details()
print("-----------------------")

s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()

print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()

print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
```

# Task 2

```python
class Assassin:
    num_of_assassin = 0
    def __init__(self,name,rate):
        self.name =name
        if rate == 100 :
            self.success_rate = rate
        else:
            self.success_rate = 100 - rate
        Assassin.num_of_assassin +=1
    @classmethod
    def failureRate(cls,name,rate):
        return cls(name,rate)

    @classmethod
    def failurePercentage(cls,name,rate):
        return cls(name,rate)

    def printDetails(self):
        print("Name:",self.name)
        print("Success rate:",self.success_rate,"%")
        print("Total number of Assassin:",Assassin.num_of_assassin)

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
class Travel:
    count = 0
    def __init__(self,source,destination):
        self.source = source
        self.destination = destination
        self.time = "1:00"
        Travel.count +=1
    def set_time(self,time):
        self.time = str(time) +":00"

    def set_source(self,source):
        self.source = source
    def set_destination(self,des):
        self.destination = des

    def display_travel_info(self):
        return  "Source: {0}\nDestination: {1}\nFlight Time: {2}".format(self.source,self.destination,self.time)

print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
from datetime import date

class Employee:
    def __init__(self, name, workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod
    @classmethod
    def employeeByJoiningYear(cls,name,startYear):
        #self.name = name
        year = date.today().year - startYear
        return cls(name,year)

    @staticmethod
    def experienceCheck(year,gender):
        isHe = ""
        if year < 3 :
            if gender == "male" :
                isHe = "He is not experienced"
            else:
                isHe = "She is not experienced"
        else:
            if gender == "male" :
                isHe = "He is experienced"
            else:
                isHe = "She is experienced"

        return isHe


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
class Cat:
    Number_of_cats = 0

    def __init__(self,color,state):
        self.color = color
        self.state = state
        Cat.Number_of_cats +=1

    @classmethod
    def no_parameter(cls):
        return cls("White","sitting")

    @classmethod
    def first_parameter(cls,color):
        return cls(color,"sitting")
    @classmethod
    def second_parameter(cls,state):
        return cls("Grey",state)

    def printCat(self):
        print(self.color,"cat is",self.state)

    def changeColor(self,color):
        self.color = color


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
import math


class Cylinder:
    radius = 5
    height = 18
    def __init__(self,r,h):
        print("Default radius =", Cylinder.radius, " and height =", Cylinder.height)
        Cylinder.radius = r
        Cylinder.height = h
        print("Updated radius =", Cylinder.radius, " and height =", Cylinder.height)

    @classmethod
    def swap(cls,r,h):
        return Cylinder(h,r)
    @classmethod
    def changeFormat(cls,st):
        r,h = st.split("-")
        return Cylinder(r,h)
    @staticmethod
    def area(r,h):
        ar = (2*math.pi*float(r)*float(r)) + (2*math.pi*float(r)*float(h))
        print("Area:",ar)
    @staticmethod
    def volume(r,h):
        vol =  math.pi*float(r)*float(r)*float(h)
        print("Volume:",vol)

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)


```

```python
class SultansDine:
    total_amount = 0
    branches = []

    def __init__(self,branch_name):
        self.branch_sell = 0.0
        self.branch_name = branch_name

    def sellQuantity(self,quantity):
        if quantity < 10 :
            self.branch_sell = quantity * 300
        elif quantity < 20 :
            self.branch_sell = quantity * 350
        else:
            self.branch_sell = quantity * 400

        SultansDine.total_amount += self.branch_sell
        SultansDine.branches.append(self)

    def branchInformation(self):
        print("Name:",self.branch_name)
        print("Branch sell:",self.branch_sell)

    @classmethod
    def details(cls):
        print("Total Number of branches:",len(SultansDine.branches))
        print("Total sell:",SultansDine.total_amount)
        for i in SultansDine.branches:
            print("Branch Name:",i.branch_name,", Branch sell:",i.branch_sell)
            print("Branch consists of total sell's:",'{0:.4g}'.format((i.branch_sell/SultansDine.total_amount)*100),"%")


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
